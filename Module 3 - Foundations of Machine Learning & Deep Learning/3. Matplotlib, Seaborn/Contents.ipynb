```python
# 1. Introduction to Matplotlib
# Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

# Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack.


# 2. Basic Plotting with Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Create a simple plot
x = np.linspace(0, 10, 100) # 100 evenly spaced values from 0 to 10
y = np.sin(x)

plt.plot(x, y)
plt.show()

```

```python
# 3. Customizing Plots
# Add labels, title, and legend
plt.plot(x, y)
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Change the line style
plt.plot(x, y, linestyle='--')
plt.show()
```

```python
# Change the line color
plt.plot(x, y, color='red')
plt.show()
```

```python
# Change the line width
plt.plot(x, y, linewidth=4)
plt.show()
```

```python
# Add markers 
plt.plot(x, y, marker='x')
plt.show()
```

```python
plt.plot(x, y)
plt.grid()
```

```python
# Save the plot
plt.savefig('plot.png')
plt.show()
```

```python
# Multiple lines on the same plot
y2 = np.cos(x)
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

```python
# Add a legend
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()   
plt.show()
```

```python
# Change the figure size
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.show()
```

```python
# 4. Plotting with Seaborn
# Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics.

import seaborn as sns

# Create a scatter plot
x = np.random.randn(100)
y = np.random.randn(100)
sns.scatterplot(x=x, y=y)
plt.show()
```

```python
# Create a line plot
sns.lineplot(x=x, y=y)
plt.show()
```

```python
# Create a bar plot
import pandas as pd
data = {'Category': ['A', 'B', 'C'], 'Values': [1, 2, 3]}
df = pd.DataFrame(data)
sns.barplot(x='Category', y='Values', data=df)
print(x)
plt.show()
```

```python
# Create a histogram
print(x)
sns.histplot(x)
plt.show()
```

```python
# Create a box plot
sns.boxplot(x)
plt.show()
```

```python
# Create a violin plot
sns.violinplot(x)
plt.show()
```

```python
# Create a heatmap
data = np.random.rand(10, 10)
sns.heatmap(data)
print(data)
plt.show()
```

```python
# Create a pair plot
data = pd.DataFrame(np.random.randn(100, 3), columns=['A', 'B', 'C'])
sns.pairplot(data)
print(data)
plt.show()
```

```python
# Create a joint plot
sns.jointplot(x='A', y='B', data=data)
plt.show()
```

```python
# Create a cat plot
data = pd.DataFrame({'x': np.random.randn(100), 'y': np.random.randn(100), 'z': np.random.choice(['A', 'B', 'C'], 100)})
sns.catplot(x='x', y='y', hue='z', data=data)
plt.show()
```
