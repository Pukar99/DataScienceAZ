```python
# 1. Variables and Data Types:
# Variables are used to store values. A variable is created the moment you first assign a value to it.
# Python has five standard data types âˆ’
# Numbers, String, List, Tuple, Dictionary

# Numbers: Python supports integers, floating point numbers and complex numbers.
num1 = 10
num2 = 20.5
num3 = 3 + 4j
print(num1, num2, num3)

# String: Strings in Python are identified as a contiguous set of characters represented in the quotation marks.
str1 = 'Hello World!'
str2 = "Python Programming"
print(str1, str2)

# List: Lists are the most versatile of Python's compound data types. A list contains items separated by commas and enclosed within square brackets ([]).
list1 = [10, 20, 30, 40, 50]
list2 = ['apple', 'banana', 'cherry']
print(list1, list2)

# Tuple: A tuple is another sequence data type that is similar to the list. A tuple consists of a number of values separated by commas.
tuple1 = (10, 20, 30, 40, 50)
tuple2 = ('apple', 'banana', 'cherry')
print(tuple1, tuple2)

# Dictionary: Python's dictionaries are kind of hash table type. They work like associative arrays or hashes found in Perl and consist of key-value pairs.
dict1 = {'name': 'John', 'age': 25, 'city': 'New York'}
print(dict1)

```

```python
# 2. Operators:
# Operators are used to perform operations on variables and values.
# Python divides the operators in the following groups:
# Arithmetic operators, Assignment operators, Comparison operators, Logical operators, Identity operators, Membership operators, Bitwise operators

# Arithmetic operators: Arithmetic operators are used with numeric values to perform common mathematical operations.
a = 10
b = 20
print(a + b)  # Addition
print(a - b)  # Subtraction
print(a * b)  # Multiplication
print(a / b)  # Division
print(a % b)  # Modulus
print(a ** b)  # Exponentiation
print(a // b)  # Floor division

# Assignment operators: Assignment operators are used to assign values to variables.
a = 10
a += 5  # a = a + 5
print(a)

# Comparison operators: Comparison operators are used to compare two values.
a = 10
b = 20
print(a == b)  # Equal
print(a != b)  # Not equal
print(a > b)  # Greater than
print(a < b)  # Less than
print(a >= b)  # Greater than or equal to
print(a <= b)  # Less than or equal to

# Logical operators: Logical operators are used to combine conditional statements.
a = 10
b = 20
print(a > 5 and b < 30)  # and
print(a > 5 or b < 30)  # or
print(not(a > 5 and b < 30))  # not

# Identity operators: Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.
a = 10
b = 20
print(a is b)  # is
print(a is not b)  # is not

# Membership operators: Membership operators are used to test if a sequence is presented in an object.
list1 = [10, 20, 30, 40, 50]
print(10 in list1)  # in
print(60 not in list1)  # not in

# Bitwise operators: Bitwise operators are used to compare (binary) numbers.
a = 10
b = 20
print(a & b)  # AND
print(a | b)  # OR
print(a ^ b)  # XOR
print(~a)  # NOT
print(a << 2)  # Zero fill left shift
print(a >> 2)  # Signed right shift


```

```python
#3. Conditional Statements:
# Conditional statements are used to perform different actions based on different conditions.
# Python supports the following conditional statements:
# if, elif, else

# if statement: Use the if statement to execute a block of code, if the condition is True.
a = 10
b = 20
if a < b:
    print('a is less than b')
    
# elif statement: Use the elif statement to specify a new condition if the first condition is False.
a = 10
b = 20
if a > b:
    print('a is greater than b')
elif a < b:
    print('a is less than b')
    
# else statement: Use the else statement to execute a block of code, if the condition is False.
a = 10
b = 20
if a > b:
    print('a is greater than b')
elif a < b:
    print('a is less than b')
else:
    print('a is equal to b')
    
# Nested if statement: You can have if statements inside if statements, this is called nested if statements.
a = 10
if a > 0:
    if a % 2 == 0:
        print('a is a positive even number')
    else:
        print('a is a positive odd number')
else:
    print('a is a negative number')
    
# Short Hand If: If you have only one statement to execute, you can put it on the same line as the if statement.
a = 10
if a > 0: print('a is a positive number')

# Short Hand If ... Else: If you have only one statement to execute, one for if, and one for else, you can put it all on the same line.
a = 10
print('a is a positive number') if a > 0 else print('a is a negative number')

# And: The and keyword is a logical operator, and is used to combine conditional statements.
a = 10
b = 20
if a > 0 and b > 0:
    print('Both a and b are positive numbers')
    
# Or: The or keyword is a logical operator, and is used to combine conditional statements.
a = 10
b = 20
if a > 0 or b > 0:
    print('Either a or b is a positive number')
    
# Nested if-else: You can have if-else statements inside if-else statements, this is called nested if-else statements.
a = 10
if a > 0:
    if a % 2 == 0:
        print('a is a positive even number')
    else:
        print('a is a positive odd number')
else:
    print('a is a negative number')
    
# Pass: if statements cannot be empty, but if you for some reason have an if statement with no content, put in the pass statement to avoid getting an error.
a = 10
if a > 0:
    pass


```

```python
# 4. Loops:
# Python has two primitive loop commands:
# while loops, for loops

# while loop: With the while loop we can execute a set of statements as long as a condition is true.
i = 1
while i <= 5:
    print(i)
    i += 1
    
# break statement: With the break statement we can stop the loop even if the while condition is true.
i = 1
while i <= 5:
    print(i)
    if i == 3:
        break
    i += 1
    
# continue statement: With the continue statement we can stop the current iteration, and continue with the next.
i = 0
while i < 5:
    i += 1
    if i == 3:
        continue
    print(i)
    
# else statement: With the else statement we can run a block of code once when the condition no longer is true.
i = 1
while i <= 5:
    print(i)
    i += 1
else:
    print('i is no longer less than or equal to 5')
    
# for loop: A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).
list1 = [10, 20, 30, 40, 50]
for i in list1:
    print(i)
    
# range() function: To loop through a set of code a specified number of times, we can use the range() function.
for i in range(5):
    print(i)
    
# break statement: With the break statement we can stop the loop before it has looped through all the items.
list1 = [10, 20, 30, 40, 50]
for i in list1:
    print(i)
    if i == 30:
        break
    
# continue statement: With the continue statement we can stop the current iteration of the loop, and continue with the next.
list1 = [10, 20, 30, 40, 50]
for i in list1:
    if i == 30:
        continue
    print(i)
    
# else statement: With the else statement we can run a block of code once when the loop is finished.
list1 = [10, 20, 30, 40, 50]
for i in list1:
    print(i)
else:
    print('Loop is finished')
    
# Nested loops: A nested loop is a loop inside a loop.
list1 = ['apple', 'banana', 'cherry']
list2 = ['red', 'yellow', 'green']
for i in list1:
    for j in list2:
        print(i, j)
        
# pass statement: for loops cannot be empty, but if you for some reason have a for loop with no content, put in the pass statement to avoid getting an error.
for i in range(5):
    pass



    
```

```python
# 7. Inheritance:
# Inheritance allows us to define a class that inherits all the methods and properties from another class.

# Parent class: The class being inherited from is called the parent class.
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname
    def printname(self):
        print(self.firstname, self.lastname)
x = Person('John', 'Doe')
x.printname()

# Child class: The class that inherits from another class is called the child class.
class Student(Person):
    pass
x = Student('Mike', 'Olsen')    
x.printname()

# Add the __init__() function: class definitions
class Student(Person):
    def __init__(self, fname, lname):
        Person.__init__(self, fname, lname)
x = Student('Mike', 'Olsen')
x.printname()

# Use the super() Function: class definitions
class Student(Person):
    def __init__(self, fname, lname):
        super().__init__(fname, lname)
x = Student('Mike', 'Olsen')
x.printname()

# Add Properties: class definitions
class Student(Person):
    def __init__(self, fname, lname, year):
        super().__init__(fname, lname)
        self.graduationyear = year
x = Student('Mike', 'Olsen', 2019)
print(x.graduationyear)

# Add Methods: class definitions
class Student(Person):
    def __init__(self, fname, lname, year):
        super().__init__(fname, lname)
        self.graduationyear = year
    def welcome(self):
        print('Welcome', self.firstname, self.lastname, 'to the class of', self.graduationyear)
x = Student('Mike', 'Olsen', 2019)
x.welcome()
```

```python
# 12. String Formatting:
# String formatting lets you inject items into a string rather than trying to chain items together using commas or string concatenation.
# As a quick comparison, consider:
player = 'Thomas'
points = 33
print('Last night, ' + player + ' scored ' + str(points) + ' points.')
# You can use string formatting to make this line a lot easier to read and write:
player = 'Thomas'
points = 33
print('Last night, %s scored %d points.' % (player, points))
# There are three ways to perform string formatting.
# The oldest method involves placeholders using the modulo % character.
# An even more convenient method uses the .format() string method.
# The newest method, introduced with Python 3.6, uses formatted string literals, called f-strings.
# Here are some examples:
player = 'Thomas'
points = 33
print('Last night, %s scored %d points.' % (player, points))
# Other examples:
a = 5
b = 10
print('Five plus ten is', a + b)
# You can basically put any Python expression inside the braces and it will be output in the string.
# Here are some examples:
name = 'John'
age = 25
print('%s is %d years old.' % (name, age))
mylist = [1, 2, 3]
print('A list: %s' % mylist)
# Here are some basic argument specifiers you should know:
# %s - String (or any object with a string representation, like numbers)
# %d - Integers
# %f - Floating point numbers
# %.<number of digits>f - Floating point numbers with a fixed amount of digits to the right of the dot.
# %x/%X - Integers in hex representation (lowercase/uppercase)
# Here are some examples:
name = 'John'
age = 25
print('Hello, %s!' % name)
print('%s is %d years old.' % (name, age))
mylist = [1, 2, 3]
print('A list: %s' % mylist)
data = ('John', 'Doe', 53.44)
format_string = 'Hello'
print('%s %s %s. Your current balance is $%s.' % (format_string, data[0], data[1], data[2]))
# Using the .format() method:
print('Hello, {}'.format('John'))
print('Hello, {}. You are {}.'.format('John', 25))
print('Hello, {}. You are {}.'.format(name, age))
print('Hello, {0}. You are {1}.'.format(name, age))
print('Hello, {1}. You are {0}.'.format(age, name))
# The .format() method has several advantages over the %s placeholder method:
# 1. You can insert objects by referring to their index, position, or key:
print('Hello, {name}. You are {age}.'.format(name='John', age=25))
person = {'name': 'John', 'age': 23}
print('Hello, {name}. You are {age}.'.format(**person))
# 2. You can use list indexing:
li = ['John', 25]
print('Hello, {0[0]}. You are {0[1]}.'.format(li))
# 3. You can use attribute access:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
p1 = Person('John', 23)
print('Hello, {0.name}. You are {0.age}.'.format(p1))
# 4. You can use element indexing:
data = ('John', 'Doe')
print('Hello, {0[0]} {0[1]}.'.format(data))
# 5. You can format numbers:
print('The number is {:d}'.format(42))
print('The number is {:f}'.format(42))
print('The number is {:b}'.format(42))
print('The number is {:o}'.format(42))
print('The number is {:x}'.format(42))
print('The number is {:X}'.format(42))
# 6. You can format dates:
import datetime
date = datetime.datetime.now()
print('The current date and time is {:%Y-%m-%d %H:%M:%S}'.format(date))
# 7. You can format percentages:
points = 19
total = 22
print('Correct answers: {:.2%}'.format(points/total))
# 8. You can round numbers:
print('The number is {:.2f}'.format(42))
# 9. You can use the string method:
name = 'John'
print('Hello, {}'.format(name))
# 10. You can use the string method with numbers:
age = 25
print('Hello, {}'.format(age))
# 11. You can use the string method with lists:
mylist = [1, 2, 3]
print('A list: {}'.format(mylist))
# 12. You can use the string method with multiple arguments:
name = 'John'
age = 25
print('Hello, {}. You are {}.'.format(name, age))
# 13. You can use the string method with dictionaries:
person = {'name': 'John', 'age': 23}
print('Hello, {name}. You are {age}.'.format(**person))
# 14. You can use the string method with list indexing:
li = ['John', 25]
print('Hello, {0[0]}. You are {0[1]}.'.format(li))



```
