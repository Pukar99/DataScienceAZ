```python
# 1. Web scraping with Beautifulsoup
# Web scraping is a technique to extract data from websites. It is a way to automate the process of copying data from websites.
# Beautifulsoup is a python library that is used to extract data from HTML and XML files.
# To install Beautifulsoup, run the following command in the terminal:
# pip install beautifulsoup4
# To install lxml, run the following command in the terminal:
# pip install lxml
# To install requests, run the following command in the terminal:
# pip install requests
# To install html5lib, run the following command in the terminal:
# pip install html5lib
# To install tweepy, run the following command in the terminal:
# pip install tweepy

```

```python
# Import the libraries
import requests
from bs4 import BeautifulSoup

# Specify the URL
url = 'https://www.crummy.com/software/BeautifulSoup/bs4/doc/'

# Connect to the website and download the HTML page
response = requests.get(url)

# Parse HTML and save to BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# To print the title of the webpage
print(soup.title)

# To print the text of the webpage
print(soup.get_text())

# To print all the links in the webpage
for link in soup.find_all('a'):
    print(link.get('href'))
    
# To print all the text in the webpage
print(soup.get_text())

```
