```python
# Insert data into the table

query = "INSERT INTO Employee (name, age, salary) VALUES (%s, %s, %s)"
values = [("Pukar", 21, 10000),
          ("Sujan", 22, 20000),
            ("Suman", 23, 30000),
            ("Sujan", 24, 40000),
            ("Suman", 25, 50000)]   
cursor.executemany(query, values)
connection.commit()
```

```python
# Fetch data from the table

cursor.execute("SELECT * FROM Employee")
result = cursor.fetchall()
for i in result:
    print(i)
```

```python
# fetching data with where clause

cursor.execute("SELECT * FROM Employee WHERE name='Pukar'")
result = cursor.fetchall()
for i in result:
    print(i)
    
```

```python
# Update data in the table

cursor.execute("UPDATE Employee SET salary=20000 WHERE name='Pukar'")
connection.commit()
```
