```python
# 1. Maximum Likelihood Estimation (MLE)
# MLE is a method for estimating the parameters of a statistical model.
# Given a dataset, we want to find the parameters that maximize the likelihood of the data.
# The likelihood is the probability of observing the data given the parameters.

# Example: Coin Toss
# Suppose we have a coin that we want to estimate the probability of heads.
# We toss the coin 10 times and get 6 heads and 4 tails.
# The likelihood of observing this data given the probability of heads is:
# P(data | p) = p^6 * (1-p)^4
# We want to find the value of p that maximizes this likelihood.

# 2. Maximum A Posteriori (MAP)
# MAP is a method for estimating the parameters of a statistical model using prior information.
# Given a dataset and a prior distribution over the parameters, we want to find the parameters that maximize the posterior distribution.
# The posterior distribution is the likelihood of the data given the parameters multiplied by the prior distribution over the parameters.

# Example: Coin Toss
# Suppose we have a prior distribution over the probability of heads that is a Beta distribution with parameters alpha=2 and beta=2.
# The posterior distribution is proportional to the likelihood times the prior:
# P(p | data) = P(data | p) * P(p)
# We want to find the value of p that maximizes this posterior distribution.

# 3. Confidence Intervals
# Confidence intervals are a way of quantifying the uncertainty in an estimate.
# Given a dataset and an estimate of a parameter, we want to find a range of values that contains the true parameter with a certain probability.
# The confidence interval is a range of values that contains the true parameter with a specified probability.

# Example: Coin Toss
# Suppose we have a dataset of coin tosses and we want to estimate the probability of heads.
# We can calculate a confidence interval for the probability of heads that contains the true value with a certain probability.

# 4. Example: Linear Regression
# Linear regression is a method for modeling the relationship between two variables.
# Given a dataset of input-output pairs, we want to find the line that best fits the data.
# The parameters of the linear regression model are the slope and intercept of the line.
# We can estimate the parameters using MLE or MAP and calculate confidence intervals for the parameters.


```
